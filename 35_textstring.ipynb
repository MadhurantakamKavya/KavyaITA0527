```python
import cv2
from google.colab.patches import cv2_imshow

def add_text_to_image(image_path, text, output_path='output_image.jpg',
                      position=(50, 50), font=cv2.FONT_HERSHEY_SIMPLEX,
                      font_scale=1, color=(255, 255, 255), thickness=2):
    """
    Adds a given text string to an image using OpenCV.

    Args:
        image_path (str): The path to the input image file.
        text (str): The text string to be added to the image.
        output_path (str): The path where the modified image will be saved.
        position (tuple): The (x, y) coordinates for the bottom-left corner of the text.
        font (int): The font type (e.g., cv2.FONT_HERSHEY_SIMPLEX).
        font_scale (float): The font scale factor.
        color (tuple): The text color in BGR format (e.g., (255, 255, 255) for white).
        thickness (int): The thickness of the text lines.
    """
    try:
        # Load the image
        image = cv2.imread(image_path)

        if image is None:
            print(f"Error: Could not load image from {image_path}")
            return

        # Add the text to the image
        cv2.putText(image, text, position, font, font_scale, color, thickness, cv2.LINE_AA)

        # Save the modified image
        cv2.imwrite(output_path, image)
        print(f"Image with text saved to {output_path}")

        # Display the modified image
        print("Displaying the image with added text:")
        cv2_imshow(image)

    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage with the provided image file
user_text = input("Please enter the text you want to appear on the image: ")
add_text_to_image('/content/image.jpg', user_text, output_path='image_with_user_text.jpg')
```
