```python
import numpy as np
from PIL import Image
from IPython.display import display # To display the image in Colab

def create_corner_boxes_image(width, height, output_filename='corner_boxes_image.png'):
    """
    Creates a white image of specified dimensions with colored boxes (Black, Blue,
    Green, Red) at each corner. The boxes are 1/10th the size of the image.

    Args:
        width (int): The width of the image in pixels.
        height (int): The height of the image in pixels.
        output_filename (str): The name of the file to save the image.

    Returns:
        PIL.Image.Image: The generated image object.
    """
    if width <= 0 or height <= 0:
        raise ValueError("Image width and height must be positive integers.")

    # Create a white image (height, width, 3 channels for RGB)
    # Each pixel value is 255 for white in 8-bit images
    image = np.ones((height, width, 3), dtype=np.uint8) * 255

    # Calculate the size of the corner boxes (1/10th of the image size)
    box_width = max(1, width // 10)  # Ensure minimum 1 pixel size
    box_height = max(1, height // 10)

    # Define RGB colors
    black = [0, 0, 0]
    blue = [0, 0, 255]
    green = [0, 255, 0]
    red = [255, 0, 0]

    # Top-Left corner (Black)
    image[0:box_height, 0:box_width] = black

    # Top-Right corner (Blue)
    image[0:box_height, (width - box_width):width] = blue

    # Bottom-Left corner (Green)
    image[(height - box_height):height, 0:box_width] = green

    # Bottom-Right corner (Red)
    image[(height - box_height):height, (width - box_width):width] = red

    # Convert the NumPy array to a PIL Image object
    pil_image = Image.fromarray(image)

    # Save the image to a file
    pil_image.save(output_filename)

    return pil_image

print("The 'create_corner_boxes_image' function has been defined.")
```

Now, let's call the function with an example size and display the generated image. You can change `image_width` and `image_height` to your desired dimensions.

```python
# Define desired image dimensions
image_width = 800
image_height = 600

# Create and save the image
generated_image = create_corner_boxes_image(image_width, image_height)

# Display the image in the Colab output
display(generated_image)

print(f"Image saved as 'corner_boxes_image.png' and displayed above.")
```
