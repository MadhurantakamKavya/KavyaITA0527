```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def create_white_image_with_rectangle(width, height, rect_start_point, rect_end_point, color=(0, 0, 0), thickness=2):
  """
  Creates a white image of a specified size and draws a rectangle on it.

  Args:
    width (int): The width of the image.
    height (int): The height of the image.
    rect_start_point (tuple): A tuple (x1, y1) representing the top-left corner of the rectangle.
    rect_end_point (tuple): A tuple (x2, y2) representing the bottom-right corner of the rectangle.
    color (tuple): The color of the rectangle in BGR format (default is black: (0, 0, 0)).
    thickness (int): The thickness of the rectangle's border. Use -1 for a filled rectangle.

  Returns:
    numpy.ndarray: The image with the drawn rectangle.
  """
  # Create a white image (255, 255, 255) with 3 color channels (BGR)
  image = np.full((height, width, 3), 255, dtype=np.uint8)

  # Draw the rectangle on the image
  cv2.rectangle(image, rect_start_point, rect_end_point, color, thickness)

  return image
```

```python
# Example usage:
# Define image dimensions
image_width = 600
image_height = 400

# Define rectangle properties
# Top-left corner (x1, y1)
rectangle_start = (100, 50)
# Bottom-right corner (x2, y2)
rectangle_end = (500, 350)
# Rectangle color (Blue in BGR format)
rectangle_color = (255, 0, 0)
# Line thickness (2 pixels)
rectangle_thickness = 2

# Create the image with the rectangle
output_image = create_white_image_with_rectangle(
    image_width, image_height,
    rectangle_start, rectangle_end,
    color=rectangle_color,
    thickness=rectangle_thickness
)

# Display the image
cv2_imshow(output_image)

# Example with a filled rectangle (red color)
output_filled_image = create_white_image_with_rectangle(
    image_width, image_height,
    (150, 100), (450, 300),
    color=(0, 0, 255),
    thickness=-1 # -1 for a filled rectangle
)

print("\nFilled Red Rectangle:")
cv2_imshow(output_filled_image)
```
