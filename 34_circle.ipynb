```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def create_white_image_with_circle(width, height, circle_center_x, circle_center_y, circle_radius, circle_color=(0, 0, 0), thickness=2):
    """
    Creates a white image of a specified size and draws a circle on it.

    Args:
        width (int): The width of the image.
        height (int): The height of the image.
        circle_center_x (int): The x-coordinate of the circle's center.
        circle_center_y (int): The y-coordinate of the circle's center.
        circle_radius (int): The radius of the circle.
        circle_color (tuple): The color of the circle in BGR format (default is black).
        thickness (int): The thickness of the circle outline. If -1, the circle is filled.

    Returns:
        numpy.ndarray: The image with the circle drawn.
    """
    # Create a white image (3 channels for BGR, 8-bit unsigned integer)
    image = np.ones((height, width, 3), dtype=np.uint8) * 255

    # Draw the circle
    center_coordinates = (circle_center_x, circle_center_y)
    image = cv2.circle(image, center_coordinates, circle_radius, circle_color, thickness)

    return image

# --- Example Usage ---
# Define image dimensions
img_width = 400
img_height = 300

# Define circle properties
circle_cx = 200
circle_cy = 150
circle_r = 70
circle_col = (255, 0, 0) # Blue color in BGR
circle_thickness = -1 # Filled circle

# Create the image with the circle
output_image = create_white_image_with_circle(
    img_width, img_height, circle_cx, circle_cy, circle_r, circle_col, circle_thickness
)

# Display the image
cv2_imshow(output_image)
print(f"Created an image of size {img_width}x{img_height} with a blue filled circle at ({circle_cx},{circle_cy}) with radius {circle_r}.")
```
